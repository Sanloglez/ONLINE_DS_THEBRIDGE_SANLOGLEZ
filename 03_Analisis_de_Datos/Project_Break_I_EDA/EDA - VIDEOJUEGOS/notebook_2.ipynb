```python
# 7. Porcentaje de valores nulos por columna
print("\nPorcentaje de valores nulos por columna:")
missing_percentage = df.isnull().sum() / len(df) * 100
print(missing_percentage.sort_values(ascending=False))
```

```python
# Eliminamos espacios antes y después de cada nombre de las columnas para poder procesarlos
df.columns = df.columns.str.strip()

# Unir todas las columnas relacionadas en una sola con separador
df['frequency'] = df[
    ['How often do you play video games?', 'How often do you play video games?']
].apply(lambda x: ', '.join(x.dropna().astype(str).replace('nan', '')), axis=1)

df['hours_per_week'] = df[
    ['How many hours do you typically spend gaming in a week?', 'How many hours do you typically spend gaming in a week?']
].apply(lambda x: ', '.join(x.dropna().astype(str).replace('nan', '')), axis=1)

df['device'] = df[
    ['Which device do you play games on the most?(Check all that apply)', 'Which device do you play games on the most?(Check all that apply)']
].apply(lambda x: ', '.join(x.dropna().astype(str).replace('nan', '')), axis=1)

df['genres'] = df[
    ['What genres of video games do you play? (Check all that apply)', 'What genres of video games do you play? (Check all that apply)']
].apply(lambda x: ', '.join(x.dropna().astype(str).replace('nan', '')), axis=1)

df['favorite_game'] = df[
    ['What is your favorite game?', 'What is your favorite game?']
].apply(lambda x: ', '.join(x.dropna().astype(str).replace('nan', '')), axis=1)

df['discovery'] = df[
    ['How do you discover new games? (Check all that apply)', 'How do you discover new games? (Check all that apply)']
].apply(lambda x: ', '.join(x.dropna().astype(str).replace('nan', '')), axis=1)

df['game_mode'] = df[
    ['Do you prefer single-player or multiplayer games?', 'Do you prefer single-player or multiplayer games?']
].apply(lambda x: ', '.join(x.dropna().astype(str).replace('nan', '')), axis=1)

df['monthly_spend'] = df[
    ['How much do you spend on gaming monthly (including in-game purchases, new games, etc.)?', 'How much do you spend on gaming monthly (including in-game purchases, new games, etc.)?']
].apply(lambda x: ', '.join(x.dropna().astype(str).replace('nan', '')), axis=1)

df['motivation'] = df[
    ['Why do you play video games? (Check all that apply)', 'Why do you play video games? (Check all that apply)']
].apply(lambda x: ', '.join(x.dropna().astype(str).replace('nan', '')), axis=1)


df[['favorite_game', 'discovery', 'motivation', 'device', 'game_mode']].head()

```
